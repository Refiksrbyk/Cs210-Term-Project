```python
import pandas as pd

df = pd.read_excel("Weight_Tracking_Final.xlsx")
df
```

```python
#adding day column to dataframe

df['Date'] = pd.to_datetime(df['Date'])

df['day'] = (df['Date'] - df['Date'].iloc[0]).dt.days + 1
```

```python
# Adding BMI to dataframe
height_m = 1.96
df['BMI'] = df['Weight'] / (height_m ** 2)
df
```
