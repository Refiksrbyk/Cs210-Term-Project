```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Weight_Tracking_altered.xlsx")

df
```

```python
sns.lineplot(x='Date', y='Pushups', data=df)
plt.xticks(rotation=45)
plt.title('Pushups Over Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Weight', 'BMI', 'Pushups']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(x='Weight', y='Pushups', data=df)
plt.title('Pushups vs Weight')
plt.show()
```

```python
df['Month'] = pd.to_datetime(df['Date']).dt.month
sns.boxplot(x='Month', y='Pushups', data=df)
plt.title('Monthly Pushups Distribution')
plt.show()
```

```python
sns.histplot(df['Pushups'], bins=30)
plt.title('Distribution of Pushup Counts')
plt.show()
```

```python
sns.regplot(x='Weight', y='Pushups', data=df)
plt.title('Regression Plot: Pushups vs Weight')
plt.show()
```
